```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df = pd.read_csv('bengaluru_house_prices.csv')
df.head()
```

```python
def extract_bhk(x):
    if isinstance(x, str):
        parts = x.split(' ')
        if len(parts) > 0:
            return int(parts[0])
    return np.nan

df['bhk'] = df['size'].apply(extract_bhk).astype('Int64')
```

```python
dropped_df = df.drop(['size','society','balcony','area_type','availability'],axis='columns')
dropped_df.head(5)
```

```python
def is_float(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
dropped_df[~dropped_df['total_sqft'].apply(is_float)].head(10)
```

```python
def convert_sqft_to_num(x):
  tokens = x.split('-')
  if len(tokens)== 2:
    return (float(tokens [0])+float(tokens[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
df4=dropped_df.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
df4.head(3)
```
